```python
from gensim.models import Word2Vec, KeyedVectors
```

```python
# Load Original Word2Vec model
word_vectors = KeyedVectors.load_word2vec_format('trmodel', binary = True)
```

```python
# Access vocabulary as a list of words
vocab_list = list(word_vectors.key_to_index.keys())
```

```python
# Access word vectors using the word as the key
word_to_vector = {word: word_vectors.get_vector(word).tolist() for word in vocab_list}

# Print the first few elements
print("First few elements of word_to_vector:")
for word, vector in list(word_to_vector.items())[:5]:
    print(f"{word}: {vector}")
```

```python
import json

# Save the word_to_vector dictionary to a JSON file
with open("word_to_vector.json", "w") as json_file:
    json.dump(word_to_vector, json_file)
```
